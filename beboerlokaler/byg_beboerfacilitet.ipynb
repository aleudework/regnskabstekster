```python
import pandas as pd 
import os
import glob

path = '/Users/alhu/Data/Konverteringsdata/Til Upload/Regnskabstekster/Beboerfaciliteter/Raw'

```

```python

# Find all Excel files in the path
excel_files = glob.glob(os.path.join(path, '*.xlsx')) + glob.glob(os.path.join(path, '*.XLSX'))

# Read each file into a DataFrame
dfs = []
for file in excel_files:
    df = pd.read_excel(file)
    dfs.append(df)

# Combine all DataFrames into one
combined_df = pd.concat(dfs, ignore_index=True)

# Konverter alle object-kolonner til strenge
for col in combined_df.select_dtypes(include=['object']).columns:
    combined_df[col] = combined_df[col].astype(str)

combined_df.to_parquet('./beboerfac.parquet')
```

```python
path2 = '/Users/alhu/VSC/regnskabstekster/beboerfac.parquet'

df2 = pd.read_parquet(path2)



print(len(df2))
```

```python
# Filtrer df2 til kun rækker hvor 'Boligtype tekst' er 'Selskabslokaler', 'Fælleshus' eller 'Fælleslokale'
df_filtered = df2[df2['Boligtype tekst'].isin(['Selskabslokaler', 'Fælleshus', 'Fælleslokale'])]

# Gruppér efter 'afdelingsnr' og 'benyttelsesart' og lav en ny df med unikke kombinationer
df_grouped = df_filtered.groupby(['Afdnr.', 'Boligtype tekst']).size().reset_index(name='antal_rækker')

# Vis den nye df
df_grouped
```

```python
import pandas as pd
import re

path = '/Users/alhu/Data/Konverteringsdata/Til Upload/Regnskabstekster/Vaskeri/Raw/19062025_Faciliteter_Alle.xlsx'
df = pd.read_excel(path)

import re

def extract_first_number(val):
    if pd.isna(val):
        return None
    s = str(val)
    m = re.search(r"\b([0-9]+)\b", s)
    return int(m.group(1)) if m else None

df_filtered = df[df['SAP Beskrivelse'].isin(['Vaskemaskine'])]

df_filtered['Afdeling'] = df_filtered['Lejemålets beliggenhed - Ej medtages'].apply(extract_first_number)

df_filtered.head()
```

```python
df_filtered.to_parquet('./vaskemaskiner.parquet', index=False)
```

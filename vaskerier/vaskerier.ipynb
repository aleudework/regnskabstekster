```python
import pandas as pd
import re

path = '/Users/alhu/Data/Konverteringsdata/Til Upload/Regnskabstekster/Vaskeri/Raw/20260213 Vaskerier.xlsx'
df = pd.read_excel(path)

def extract_afdeling(val):
    if pd.isna(val):
        return None
    m = re.search(r"(?:\\t|\t)\s*([0-9]+)", str(val))
    return int(m.group(1)) if m else None

df['Afdeling_SAP'] = df['Afdeling'].astype(str).apply(extract_afdeling)

df2 = df[['Afdeling_SAP']]
df2.to_parquet('./vaskerier.parquet', index=False)
```
